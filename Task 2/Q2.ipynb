```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/SiddharthS/Downloads/best-fit.csv")
x=data['X'].values
y=data['Y'].values
```

```python
m,c=np.polyfit(x,y,1)
regline=m*x+c
regerror=np.sum((regline-y.mean())**2)
toterror=np.sum((y-y.mean())**2)
r2=regerror/toterror
```

```python
plt.scatter(x,y)
plt.plot(x,regline,color='black', linewidth=2)
equation=f'Equation: y={m:.4f}x+{c:.4f}'
r2_show=f'R^2={r2:.4f}'
plt.text(x.min(),y.max(),equation)
plt.text(x.min(),y.max()-0.1*y.mean(),r2_show)
plt.title('Best Fit Line')
plt.show()
```
